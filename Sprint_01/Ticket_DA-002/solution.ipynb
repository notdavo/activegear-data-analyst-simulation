```python
import pandas as pd
```

```python
#Task 1
products = pd.read_csv('../../datasets/raw/products.csv')
orders = pd.read_csv('../../datasets/raw/orders.csv')
order_items = pd.read_csv('../../datasets/raw/order_items.csv')
```

```python
#Task 2
completed_orders = orders[orders['status'] == 'Completed'].copy()
completed_orders
```

```python
products
```

```python
#Task 3
full_orders = completed_orders.merge(right=order_items, how='left', on='order_id').copy()
full_orders
```

```python
#Task 4
orders_with_products = full_orders.merge(products, 'left', 'product_id')
orders_with_products
```

```python
#Task 5
orders_with_products["net_amount"] = (orders_with_products["quantity"] * orders_with_products["unit_price"] * (1 - orders_with_products["discount_pct"] / 100))
orders_with_products
```

```python
#Task 6 and 7
grouped_products = orders_with_products.groupby(['product_id', 'product_name', 'category', 'subcategory']).agg(units_sold = ('quantity', 'sum'), order_count = ('order_id','nunique'), total_revenue=('net_amount','sum')).reset_index()
grouped_products
```

```python
#Task 8
grouped_products.sort_values(by='total_revenue', ascending=False)
grouped_products
```

```python
#Task 9
grouped_products.to_csv('../../output/product_sales_performance.csv')
```
