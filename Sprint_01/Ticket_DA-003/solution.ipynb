```python
import pandas as pd
```

```python
#Task 1 Load stores, orders and order_items
stores = pd.read_csv('../../datasets/raw/stores.csv')
orders = pd.read_csv('../../datasets/raw/orders.csv')
order_items = pd.read_csv('../../datasets/raw/order_items.csv')
```

```python
orders
```

```python
order_items
```

```python
stores
```

```python
#Task 2 Keep only completed orders
completed_orders = orders[orders['status'] == 'Completed'].copy()
completed_orders
```

```python
#Task 3 Create net_amount using quantity, unit_price and discount_pct
order_items['net_amount'] = (order_items['quantity'] * order_items['unit_price'] * (1 - order_items['discount_pct'] / 100))
order_items
```

```python
#Task 4 Merge order_items with orders
orders_with_items = order_items.merge(completed_orders, how='left', on='order_id')
orders_with_items
```

```python
#Task 5 Merge the result with stores
orders_with_items_stores = orders_with_items.merge(stores, how='left', on='store_id')
orders_with_items_stores
```

```python
#Task 6 Group by order_date, store_id, store_name, country and city 
#Task 7 Calculate total_orders, total_items_sold and total_revenue
grouped_order_full_data = orders_with_items_stores.groupby(['order_date','store_id','store_name','country','city']).agg(total_orders = ('order_id', 'nunique'), total_items_sold = ('quantity', 'sum'), total_revenue = ('net_amount', 'sum'))
grouped_order_full_data
```

```python
#Task 8 Sort by order_date ascending and total_revenue descending
grouped_order_full_data = grouped_order_full_data.reset_index().sort_values(['order_date', 'total_revenue'], ascending=[True,False])
grouped_order_full_data
```

```python
#Task 9 Export the result as output/store_daily_sales.csv
grouped_order_full_data.to_csv('../../output/store_daily_sales.csv', index=False)
```
